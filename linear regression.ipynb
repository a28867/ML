```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array([[6],[8],[10],[14],[18]])
y = [7,9,13,17.5,18]
```

```python
plt.axis([0,25,0,25])
plt.grid(True)
plt.plot(X,y,'.');
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
pizza1 = np.array([[12]])
predicted = model.predict(pizza1)
```

```python
predicted
```

```python
plt.axis([0,25,0,25])
plt.grid(True)
plt.plot(X,y,'--')
plt.plot(pizza1,predicted,'.');
```

```python
test_pizza = np.array([[12],[20]])
predicted_pizza = model.predict(test_pizza)
```

```python
plt.axis([0,25,0,25])
plt.grid(True)
plt.plot(X,y,'--')
plt.plot(test_pizza,predicted_pizza,'.');
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
plt.axis([0,25,0,25])
plt.grid(True)
plt.plot(X,y,'.')
plt.plot(X,model.predict(X))
plt.plot(test_pizza,predicted_pizza,'r.');
```

```python
XL = np.array([[0],[25]])
plt.axis([0,25,0,25])
plt.grid(True)
plt.plot(X,y,'.')
plt.plot(XL,model.predict(XL))
plt.plot(test_pizza,predicted_pizza,'r.');
```

```python
XL = np.array([[0],[25]])
plt.title('Pizza price')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.axis([0,25,0,25])
plt.grid(True)
plt.plot(X,y,'.')
plt.plot(XL,model.predict(XL))
plt.plot(test_pizza,predicted_pizza,'r.');
```

```python
XL = np.linspace(0,25,100)
XL
```

```python
print(XL.reshape(-1,1))
```

```python
plt.title('Pizza price')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.axis([0,25,0,25])
plt.grid(True)
plt.plot(X,y,'.')
plt.plot(XL,model.predict(XL.reshape(-1,1)))
plt.plot(test_pizza,predicted_pizza,'r.');
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
f1 = PolynomialFeatures(degree=2)
XX = f1.fit_transform(X)
```

```python
XX
```

```python
quadratic_model = LinearRegression()
quadratic_model.fit(XX,y)
```

```python
X_quadratic = f1.transform(np.linspace(0,25,100).reshape(-1,1))
X_quadratic
```

```python
plt.title('Pizza price')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.axis([0,25,0,25])
plt.grid(True)
plt.plot(X,y,'.')
plt.plot(XL,model.predict(XL.reshape(-1,1)))
plt.plot(XL,quadratic_model.predict(X_quadratic))
plt.plot(test_pizza,predicted_pizza,'r.');
```

```python
quadratic_test_pizza = f1.transform(test_pizza)
quadratic_test_pizza
```

```python
plt.title('Pizza price')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.axis([0,25,0,25])
plt.grid(True)
plt.plot(X,y,'.')
plt.plot(XL,model.predict(XL.reshape(-1,1)))
plt.plot(XL,quadratic_model.predict(X_quadratic))
plt.plot(test_pizza,predicted_pizza,'r.')
plt.plot(test_pizza,quadratic_model.predict(quadratic_test_pizza),'rD');
```

```python
quadratic_model.coef_
```

```python
quadratic_model.intercept_
```

```python
X_test = [[8],[9],[11],[16],[12]]
y_test = [11,8.5,15,18,11]
```

```python
model.score(X_test,y_test)
```

```python
quadratic_model.score(f1.transform(X_test),y_test)
```
